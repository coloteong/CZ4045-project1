```python
noun_adj_pairs = []

def checkconj(token, noun):
    for child in token.children:
        if child.dep == conj:
            noun_adj_pairs.append((noun,child.lemma_))
            checkconj(child, noun)
                 
for rating in rating1:
    sentence_spans = list(rating.sents)
    for sentence in sentence_spans:
        for token in sentence:
            if token.dep == amod and token.head.pos is NOUN:
                noun = token.head.lemma_ 
                noun_adj_pairs.append((noun,token.lemma_))
                #check if there are other ADJ conjuncted 
                checkconj(token, noun)
                
            elif token.dep == acomp:
                #print(token.text)
                for child in token.head.children:
                    if(child.dep == nsubj):
                        noun = child.lemma_ if token.head.pos == NOUN else child.text
                        noun_adj_pairs.append((noun, token.lemma_))
                        #check if there are other ADJ conjuncted 
                        checkconj(token, noun)
noun_adj_pairs
```

## 2. Getting adjective and noun pair usin dependencies 

```python
from spacy.symbols import nsubj, amod, nsubj, acomp, VERB, NOUN, AUX
from collections import Counter
```

```python
rating1 = reviews.loc[reviews['stars'] == 1]
rating1 = rating1.sample(50)
rating1 = list(rating1['text'])
rating1 = [nlp(rating) for rating in rating1]
```

```python
noun_adj_pairs = []
for rating in rating1:
    sentence_spans = list(rating.sents)
    for sentence in sentence_spans:
        for possible_subject in sentence:
            if possible_subject.dep == amod and possible_subject.head.pos == NOUN:
                noun_adj_pairs.append((possible_subject.head.text.lower(),possible_subject.text.lower()))
            elif possible_subject.dep == acomp:
                #print(possible_subject.text)
                for child in possible_subject.head.children:
                    if(child.dep == nsubj):
                        noun_adj_pairs.append((child.text.lower(), possible_subject.text.lower()))
noun_adj_pairs
```

```python
c = Counter(noun_adj_pairs)
print (c.most_common(10))
```
